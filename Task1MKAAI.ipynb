```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated()
```

```python
sns.countplot(data = df , x=df.Attrition, order=df['Attrition'].value_counts().index,hue=df.Department,palette=['#a1c9f4', '#8de5a1', '#ff9f9b'])
```

```python
meanSalary=df['MonthlyIncome'].mean()
meanSalary
```

```python
salaryLevel=[]
for i in df['MonthlyIncome']:
    if i < meanSalary:
      salaryLevel.append('below average')
    else:
      salaryLevel.append ('above average')
```

```python
df['SalaryLevel'] = salaryLevel
df
```

```python
salary_attire = df.groupby('SalaryLevel')['Attrition'].value_counts().reset_index()
salary_attire
```

```python
sns.barplot(data=salary_attire, x='SalaryLevel', y='count', hue='Attrition',palette=['#98a3d0','#36214c'])
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['AttritionEncoded'] = encoder.fit_transform(df['Attrition'])
df['AttritionEncoded'] #1=yes 2=No
```

```python
sns.heatmap(df[['JobSatisfaction','AttritionEncoded']].corr(),cmap="crest",annot=True)
plt.show()
```

```python
df['YearsSinceLastPromotion'].unique()
```

```python
promotion_attire = df.groupby('YearsSinceLastPromotion')['AttritionEncoded'].mean() * 100
promotion_attire
```

```python
plt.figure(figsize=(8,5))
plt.plot(promotion_attire.index, promotion_attire.values, marker='o')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Attrition Rate (%)')
plt.show()
```
